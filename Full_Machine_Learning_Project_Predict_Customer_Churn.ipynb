##Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train.drop(columns=['customerID']))
X_test_scaled = scaler.transform(X_test.drop(columns=['customerID']))
```

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression()
```

```python
model_1.fit(X_train_scaled,y_train)
```

```python
y_pred = model_1.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score , f1_score, confusion_matrix
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
preci = precision_score(y_test,y_pred)
```

```python
preci
```

```python
recall = recall_score(y_test,y_pred)
recall
```

```python
f1 = f1_score(y_test,y_pred)
```

```python
f1
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
##Model 2
from sklearn.tree import DecisionTreeClassifier
model_2 = DecisionTreeClassifier()
```

```python
model_2.fit(X_train_scaled,y_train)
```

```python
y_pred2 = model_2.predict(X_test_scaled)
```

```python
acc_dt = accuracy_score(y_test,y_pred2)
preci_dt = precision_score(y_test,y_pred2)
recall_dt = recall_score(y_test,y_pred2)
f1_dt = f1_score(y_test,y_pred2)
cm_dt = confusion_matrix(y_test,y_pred2)

print(f"Decision Tree Accuracy: {acc_dt}")
print(f"Decision Tree Precision: {preci_dt}")
print(f"Decision Tree Recall: {recall_dt}")
print(f"Decision Tree F1-Score: {f1_dt}")
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues')
plt.title('Decision Tree Confusion Matrix')
```
